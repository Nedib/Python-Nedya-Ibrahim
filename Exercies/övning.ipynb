Simulations of one die,

````python
import numpy as np
import matplotlib.pyplot as plt

# Number of sides on the dice
num_sides = 6

# Theoretical mean of a six-sided dice
theoretical_mean = (1 + num_sides) / 2

# Number of simulations for each roll count
roll_counts = [10, 100, 1000, 10000, 100000, 1000000, 10000000]

# Initialize a list to store means for each roll count
means = []

# Simulate rolling dice and calculate means
for rolls in roll_counts:
    dice_rolls = np.random.randint(1, num_sides + 1, rolls)
    mean = np.mean(dice_rolls)
    means.append(mean)

# Plot the means
plt.plot(roll_counts, means, marker='o')
plt.axhline(theoretical_mean, color='red', linestyle='--', label='Theoretical Mean')
plt.title("Mean of Dice Rolls vs. Number of Rolls")
plt.xlabel("Number of Rolls")
plt.ylabel("Mean")
plt.xscale('log')
plt.legend()
plt.grid()
plt.show()
````

## 1. Step tracker

Create a class that can be used as a step tracker. It should have a property "steps" that is read only; a method step() that increases "steps" by 1 each time it is called; and a method reset() that resets the counter.

Instantiate the class, and write a loop that simulates walking 1000 steps. Then print the value of "steps".

````python
class StepTracker:
    def __init__(self):
        self._steps = 0

    @property
    def steps(self):
        return self._steps

    def step(self):
        self._steps += 1

    def reset(self):
        self._steps = 0

tracker = StepTracker()

for _ in range(1000):
    tracker.step()

print("Total steps walked:", tracker.steps)
````

````python
"""Självklart! Här är förklaringen av koden med avseende på de fyra grundpelarna i objektorienterad programmering (OOP):
u
Inkapsling:

Inkapsling handlar om att gömma och skydda egenskaper och metoder för att förhindra oönskad åtkomst eller ändring från andra delar av koden.
I koden har vi använt inkapsling genom att lägga till ett understreck före egenskapen "_name" (till exempel "_name"). 
Detta indikerar att den är en privat egenskap och bör inte ändras direkt från andra delar av koden.
Abstraktion:

Abstraktion innebär att visa endast det som är viktigt och nödvändigt för användaren,
 medan tekniska detaljer göms bort.
I vår kod har vi skapat en överordnad klass "Animal" som innehåller en abstrakt metod "make_sound".

 Detta gör att vi kan fokusera på djurens förmåga att göra ljud utan att bry oss om detaljerna om hur ljudet faktiskt skapas.
Arv:

Arv möjliggör återanvändning av kod och skapande av hierarkier av klasser där en underklass (child class) ärver egenskaper och metoder från en överordnad klass (parent class).
I vår kod har vi skapat två underklasser, "Cat" och "Bird", som ärver från den överordnade klassen "Animal". Detta innebär att de har tillgång till alla egenskaper och metoder som är definierade i "Animal".
Polymorfism:

Polymorfism möjliggör att olika klasser kan använda samma metod på olika sätt. Det innebär att klasser kan ha samma namngivna metoder, men bete sig olika beroende på sin specifika implementering.
I vår kod har både "Cat" och "Bird" en metod med namnet "make_sound". Trots att de har samma metodnamn, beter de sig olika när de anropas. Katten gör ett mjau-ljud medan fågeln kvittrar.
Sammanfattningsvis använder vår kod objektorienterade principer, inklusive inkapsling för att skydda egenskaper, abstraktion för att förenkla representationen av djur, arv för att organisera klasser hierarkiskt och polymorfism för att olika klasser kan dela metodnamn men utföra olika handlingar när de anropas.
 Detta gör koden mer läsbar, underhållbar och flexibel."""




# Inkapsling: Lås in egenskaperna i klasserna för att skydda dem från extern åtkomst.

# Abstraktion: Fokusera bara på de viktigaste egenskaperna och beteendena i klasserna.

# Arv: Skapa en överordnad klass (Animal) som innehåller gemensamma egenskaper och låt andra klasser ärva från den.

# Polymorfism: Låt klasserna (katter och fåglar) använda samma metod (make_sound) på olika sätt.

# Överordnad klass Animal
class Animal:
    def __init__(self, name):
        self._name = name  # Inkapslad egenskap: namn

    # Abstraktion: Metod för att göra ljud
    def make_sound(self):
        pass  # Abstrakt metod som implementeras i underklasser

# Underklass Cat som ärver från Animal
class Cat(Animal):
    def make_sound(self):
        return "Mjau!"  # Katten gör ett mjau-ljud

# Underklass Bird som ärver från Animal
class Bird(Animal):
    def make_sound(self):
        return "Kvitt, kvitt!"  # Fågeln gör ett kvittrande ljud

# Skapa instanser av katter och fåglar
cat1 = Cat("Whiskers")
cat2 = Cat("Mittens")
cat3 = Cat("Felix")
bird1 = Bird("Tweetie")
bird2 = Bird("Chirpy")

# Anropa make_sound-metoden för varje djur
animals = [cat1, cat2, cat3, bird1, bird2]
for animal in animals:
    print(f"{animal._name} säger: {animal.make_sound()}")

# Lås egenskapen "_name" och förhindra direkt åtkomst från andra delar av koden
# animal._name = "New Name"  # Ger felmeddelande på grund av inkapslingen
````

OOP (Objektorienterad programmering): OOP är en programmeringsparadigm som använder objekt och klasser för att organisera och strukturera kod. Objekt är instanser av klasser, och de kan innehålla både data (egenskaper) och funktioner (metoder) som kan utföra olika uppgifter. OOP främjar koncept som inkapsling, abstraktion, arv och polymorfism för att skapa återanvändbar och modulär kod.

Properties: Properties är speciella metoder i Python som används för att styra åtkomsten till klassens egenskaper (variabler). De ger en flexibel och säker metod för att läsa och skriva data i objekt. Med properties kan du sätta gränser och regler för hur egenskaperna hanteras.

Dunder Methods (Dunder metoder): Dunder-metoder är specialmetoder i Python som har dubbla understreck i sina namn, t.ex., __init__, __str__, __add__. Dessa metoder styr hur objekt av en klass ska bete sig i olika situationer och ger oss möjlighet att anpassa standardbeteendet för våra klasser. Exempelvis används __init__ för att initialisera objekt, och __str__ för att anpassa hur objekt representeras som strängar.

User Exceptions (Användardefinierade undantag): I Python kan du skapa dina egna undantag (exception) genom att ärva från inbyggda undantag eller undantag som ärver från Exception-klassen. Användardefinierade undantag används när du vill signalera och hantera specifika fel eller situationer i din kod.

Inheritance (Arv): Arv är ett OOP-koncept som tillåter en klass att ärva egenskaper och metoder från en annan överordnad klass. Detta främjar återanvändbarhet och möjliggör hierarkier av klasser. En underklass (subklass) kan utöka eller modifiera funktionaliteten som ärvts från en överordnad klass.

Operator Overloading (Överbelastning av operatorer): Python tillåter överbelastning av operatörer som +, -, *, etc., genom att definiera speciella metoder som __add__, __sub__, __mul__, etc. Dessa metoder styr hur operatorer ska fungera med objekt av din klass. Till exempel, om du överlagrar __add__, kan du anpassa hur två objekt adderas tillsammans med +.

Dessa begrepp är grundläggande inom programmering och objektorienterad programmering (OOP) och är nyckeln till att skapa effektiv och läsbar kod i Python och många andra programmeringsspråk.



Getter Method (Get-metod): En getter-metod är en speciell metod i en klass som används för att hämta eller läsa en privat egenskap (variabel) från objektet. Den kallas vanligtvis med ett enkelt attribut och returnerar värdet på den privata egenskapen. Getter-metoder används ofta i samband med inkapsling för att ge kontrollerad åtkomst till egenskaper och möjliggöra validering eller extra logik vid åtkomst.
