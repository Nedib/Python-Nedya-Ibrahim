## 1. Dice rolls (*)

Simulate 10 dice rolls and **append** the rolls to a list or use **list comprehension**.

&nbsp; a) &nbsp; sort the list in **ascending** order (*)

&nbsp; b) &nbsp; sort the list in **descending** order (*)

&nbsp; c) &nbsp; find the **maximum** and **minimum** value in the list (*)


<details>

<summary>Answer </summary>

```

```

````python
import random

#simulate 10 dice rolls and append them to a list
dice_rolls = [random.randint(1, 6,) for _ in range(10)]

#sort the list in asceding order
ascending_order = sorted(dice_rolls)

#sort the list in descending order
descending_order = sorted(dice_rolls, reverse=True)

#Find the maximum and minimum values in the list
maximum_value = max (dice_rolls)
minimum_value = min (dice_rolls)

#Print the results
print("Dice Rolls: ", dice_rolls)
print("Sorted in Ascending Order:", ascending_order)
print("Sorted in Descending Order:", descending_order)
print("Maximum Value:", maximum_value)
print("Minimum Value:", minimum_value)
````

<img align="left" src="../Assets/probability_six_graph.png" width="300"/>

````python
import random
import matplotlib.pyplot as plt

def simulate_dice_rolls(num_rolls):
    outcomes = [random.randint(1, 6) for _ in range(num_rolls)]
    count_six = outcomes.count(6)
    probability_six = count_six / num_rolls
    return count_six, probability_six

def plot_probability_six(probability_six, num_rolls):
    plt.plot(probability_six, '-*')
    plt.title("Probability of Six for Different Numbers of Rolls")
    plt.xticks(range(len(num_rolls)), num_rolls)
    plt.xlabel("Number of Dice Rolls")
    plt.ylabel("Probability")
    plt.show()

if __name__ == "__main__":
    num_rolls = [10, 100, 1000, 10000, 100000, 1000000]
    count_six_list = []
    probability_six = []

    for rolls in num_rolls:
        count_six, prob_six = simulate_dice_rolls(rolls)
        count_six_list.append(count_six)
        probability_six.append(prob_six)

    for i, rolls in enumerate(num_rolls):
        print(f"Number of outcome six in {rolls} dice rolls: {count_six_list[i]}")
        print(f"Probability of outcome six: {probability_six[i]:.4f}\n")

    plot_probability_six(probability_six, num_rolls)
````

## 6. Monte Carlo simulation (**)

&nbsp; a) Simulate 5000 uniformly distributed points between -1 and 1, $X\sim U(-1,1)$. Now calculate the Euclidean distances between each point and the origin (0,0), $d_i = \sqrt{x_i^2 + y_i^2}$. For $d_i < 1$ plot the point with one color, and for $d_i > 1$ plot the point in another color, $i = \{1,\ldots, 5000\}$.  (**)

&nbsp; b) Calculate the fraction between number of inner points and number of outer points. What do you this will converge to when number of points reaches infinity? (**)

````python
import random
import math
import matplotlib.pyplot as plt

def monte_carlo_simulation(num_points):
    inner_points = []
    outer_points = []

    for _ in range(num_points):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        distance = math.sqrt(x ** 2 + y ** 2)

        if distance < 1:
            inner_points.append((x, y))
        else:
            outer_points.append((x, y))

    return inner_points, outer_points

def calculate_fraction(inner_points, total_points):
    return len(inner_points) / total_points

def plot_monte_carlo_results(inner_points, outer_points):
    inner_x, inner_y = zip(*inner_points)
    outer_x, outer_y = zip(*outer_points)

    plt.figure(figsize=(6, 6))
    plt.scatter(inner_x, inner_y, color='blue', marker='o', s=5, label='Inner Points')
    plt.scatter(outer_x, outer_y, color='red', marker='o', s=5, label='Outer Points')
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.legend()
    plt.title("Monte Carlo Simulation")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()

if __name__ == "__main__":
    num_points = 5000
    inner_points, outer_points = monte_carlo_simulation(num_points)
    total_points = num_points

    fraction = calculate_fraction(inner_points, total_points)
    print(f"Fraction (inner/total): {fraction:.4f}")

    plot_monte_carlo_results(inner_points, outer_points)
````

## 7. A cute rabbit among two ferocious snakes (**)

There are three identical doors, behind two of them, there are ferocious snakes and behind one of them is a cute little rabbit. Your aim is to pick a door and get the cute rabbit. You start picking one door randomly and suddenly another door opens and snake slithers away *pssss*. Now you have a choice to either stay with your door or change to the remaining door. 

<img src="../Assets/snakes_rabbits.png" width = 200>

&nbsp; a) What is your choice and why? (**)

&nbsp; b) Randomly assign which door has the rabbit, and randomly pick an initial door. Now simulate 10, 100, 1000, 10000, 100000, 1000000 times for staying and for switching doors. Calculate the proportion of finding the rabbits for each set of simulations. Plot the results (**)


<details>

````python
import random
import matplotlib.pyplot as plt

def simulate_monty_hall(num_simulations):
    stay_wins = 0
    switch_wins = 0
    stay_proportions = []
    switch_proportions = []

    for _ in range(num_simulations):
        # Randomly assign which door has the rabbit (0, 1, or 2)
        rabbit_door = random.randint(0, 2)
        
        # Randomly pick an initial door (0, 1, or 2)
        initial_choice = random.randint(0, 2)
        
        # Monty Hall opens one of the other doors with a snake behind it
        other_doors = [door for door in range(3) if door != initial_choice]
        opened_door = random.choice([door for door in other_doors if door != rabbit_door])
        
        # Calculate the other remaining door
        remaining_door = [door for door in range(3) if door != opened_door and door != initial_choice][0]

        # Check if staying with the initial choice results in finding the rabbit
        if initial_choice == rabbit_door:
            stay_wins += 1
        
        # Check if switching doors results in finding the rabbit
        if remaining_door == rabbit_door:
            switch_wins += 1
        
        stay_proportions.append(stay_wins / (num_simulations))
        switch_proportions.append(switch_wins / (num_simulations))
    
    return stay_proportions, switch_proportions

def plot_monty_hall_results(stay_proportions, switch_proportions):
    plt.plot(stay_proportions, label="Stay")
    plt.plot(switch_proportions, label="Switch")
    plt.xlabel("Number of Simulations")
    plt.ylabel("Proportion of Finding the Rabbit")
    plt.legend()
    plt.title("Monty Hall Simulation Results")
    plt.show()

if __name__ == "__main__":
    num_simulations = 1000000
    stay_proportions, switch_proportions = simulate_monty_hall(num_simulations)
    plot_monty_hall_results(stay_proportions, switch_proportions)
````
